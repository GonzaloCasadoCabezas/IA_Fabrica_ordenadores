```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración visualización
%matplotlib inline
plt.style.use('ggplot')
sns.set_palette("husl")
```

```python
df = pd.read_csv("./simulation_results_20250404_134637.csv")
```

```python
print("Dimensiones del dataset:", df.shape)
print("\nPrimeras filas:")
display(df.head())

# Estadisticas descriptivas
print("\nEstadisticas descriptivas:")
display(df.describe(include='all').round(2))
```

```python
# Tiempos de procesamiento
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Component', y='Total_Time')
plt.title('Distribución de tiempos totales por componente')
plt.xticks(rotation=45)
plt.ylabel('Tiempo Total')
plt.show()
```

```python
# Retrasos en transporte
plt.figure(figsize=(10,5))
sns.histplot(df['Transport_Delay'], bins=20, kde=True)
plt.title('Distribución de retrasos en trasnsporte')
plt.xlabel('Retraso (unidades de tiempo)')
plt.show()
```

```python
components_stats = df.groupby('Component').agg({'Total_Time': ['mean', 'std', 'min', 'max'], 'Transport_Delay': ['mean', 'median'], 'Status': lambda x: (x == 'Success').mean()}).round(2)

components_stats.columns = ['_'.join(col).strip() for col in component_stats.columns.values]
components_stats = components_stats.rename(columns={'Status_<lambda>': 'Success_rate'})

print("Estadísticas por componentes:")
display(components_stats.sort_values('Total_Time_mean', ascending = False))
```
